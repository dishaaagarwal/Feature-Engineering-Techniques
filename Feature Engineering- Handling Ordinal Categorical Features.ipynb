Count or Frequency Encoding

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
train_set.head()
```

```python
columns=[1,3,5,6,7,8,9,13]
```

```python
train_set=train_set[columns]
train_set.head()
```

```python
train_set.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
train_set.head()
```

```python
for features in train_set.columns:
    print(features,":",len(train_set[features].unique())," labels")
```

```python
countrymap=train_set['Country'].value_counts().to_dict()

train_set['Country']=train_set['Country'].map(countrymap)
train_set.head()
```

#### Advantages

1. Easy to Implement

2. Not increasing the Feature Space

#### Disadvantages

1. If both the labels have same frequency then it will consider both the labels as same
   
   eg- Suppose INDIA -5 and USA-5 then it will consider both the label as same.

####  Target Guided Label Encoding

1.Ordering the labels according to the target


2.Replace the labels by the joint probability of being 1 or 0

```python
import pandas as pd
df=pd.read_csv('titanic.csv', usecols=['Cabin','Survived'])
df.head()
```

```python
df.isnull().sum()
```

```python
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Cabin_Encoding']=df['Cabin'].astype(str).str[0]
df.head()
```

```python
df['Cabin_Encoding'].unique()
```

```python
#We have taken the labels(CABIN) according to the Target Label(SURVIVED)
target_ordinal_labels=df.groupby('Cabin_Encoding')['Survived'].mean().sort_values().index
target_ordinal_labels
```

```python
ordinal_labels2={k:i for i,k in enumerate(target_ordinal_labels,0)}
ordinal_labels2
```

```python
df['Cabin_Target_ordinal_labels']=df['Cabin_Encoding'].map(ordinal_labels2)
df.head()
```
